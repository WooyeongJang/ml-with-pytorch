```python
# 데이터셋
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
#훈련 데이터셋, 테스트 데이셋
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, stratify= y)
```

```python
# 전처리 (훈련과 테스트)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
# 퍼셉트론 모델을 불러와서 훈련데이터로 학습
from sklearn.linear_model import Perceptron
model0 = Perceptron(eta0 = 0.1 , random_state=1)
model0.fit(X_train_std, y_train)

# 회구
```

```python
y_pred = model0.predict(X_test_std)
```

```python
print("테스트 샘플", len(y_pred))
print("오분류갯수",(y_test != y_pred).sum())
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
model0.score(X_test_std, y_test)
```

```python
# 데이터셋
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target

#훈련 데이터셋, 테스트 데이셋
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, stratify= y)

# 전처리 (훈련과 테스트)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
# 모델 비교를위해 라이브러리 가져오기
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import Perceptron
```

```python
#퍼셉트론 비교
model11 = Perceptron(eta0 = 0.1 , random_state=1)
model12 = SGDClassifier(loss='perceptron', eta0=0.1, random_state=1)
model11.fit(X_train_std, y_train)
model12.fit(X_train_std, y_train)

# 로지스틱 회귀 비교
model21 = LogisticRegression(C=1000.0, random_state=1)
model22 = SGDClassifier(loss='log_loss', random_state=1)
model21.fit(X_train_std, y_train)
model22.fit(X_train_std, y_train)

# SVM 비교
model31 = SVC(kernel='linear', C=1.0, random_state=1)
model32 = SGDClassifier(loss='hinge', random_state=1)
model31.fit(X_train_std, y_train)
model32.fit(X_train_std, y_train)
```
