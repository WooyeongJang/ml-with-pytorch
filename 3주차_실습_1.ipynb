```python
from sklearn.impute import SimpleImputer
import numpy as np
imr = SimpleImputer(missing_values=np.nan, strategy='mean')
imr = imr.fit(df.values)
```

```python
imputed_data = imr.transform(df.values)
imputed_data
```

```python
# median
imr_median = SimpleImputer(missing_values=np.nan, strategy='median')
imr_median = imr.fit(df.values)
imputed_data_median = imr_median.transform(df.values)
# most_frequent
imr_most_frequent = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imr_most_frequent = imr_most_frequent.fit(df.values)
imputed_data_median_most_frequent = imr_most_frequent.transform(df.values)
# constant
imr_constant = SimpleImputer(missing_values=np.nan, strategy='constant')
imr_constant = imr.fit(df.values)
imputed_data_median_constant = imr_constant.transform(df.values)
```

```python
print(imputed_data_median)
print(imputed_data_median_most_frequent)
print(imputed_data_median_constant)
```

```python
from sklearn.preprocessing import FunctionTransformer
ftr_imr = FunctionTransformer(lambda X: imr.fit_transform(X.T).T, validate=False)
imported_data = ftr_imr.fit_transform(df.values)
imported_data
```

```python
from sklearn.preprocessing import FunctionTransformer

# 1. 함수 정의 (lambda 대신)
def impute_rowwise(X):
    # 행 기준 처리를 위해 전치 → imputer 적용 → 다시 전치
    return imr.fit_transform(X.T).T

# 2. FunctionTransformer에 함수 전달
ftr_imr = FunctionTransformer(impute_rowwise, validate=False)
# 3. 데이터 적용
imported_data = ftr_imr.fit_transform(df.values)
# 4. 결과 출력
print(imported_data)
```

```python
# 4.2.1
import pandas as pd

df = pd.DataFrame([['green', 'M', 10.1, 'class2'],
                   ['red', 'L', 13.5, 'class1'],
                   ['blue', 'XL', 15.3, 'class2']])

df.columns = ['color', 'size', 'price', 'classlabel']
df
```

```python
# 4.2.2 순서가 있는 특성 매핑
# XL = L +1 = M + 2
size_mapping = {'XL': 3,
                'L': 2,
                'M': 1}

df['size'] = df['size'].map(size_mapping)
df
```

```python
inv_size_mapping = {v: k for k, v in size_mapping.items()}
df['size'].map(inv_size_mapping)
```

```python
# 4.2.3 클래스 레이블 인코딩
# 1. 클래스 라벨 가져오기
labels = df['classlabel']
# 2. 중복 제거 (고유값만 추출)
unique_labels = list(set(labels))
# 3. 정렬 (순서 고정)
unique_labels.sort()
# 4. 빈 딕셔너리 만들기
class_mapping = {}
# 5. 하나씩 숫자 붙이기
idx = 0
for label in unique_labels:
    class_mapping[label] = idx
    idx += 1
# 결과
print(class_mapping)
```

```python
# 4.2.3 클래스 레이블 인코딩
import numpy as np
class_mapping = {label:idx for idx, label in enumerate(np.unique(df['classlabel']))}
class_mapping
```

```python
from sklearn.preprocessing import LabelEncoder

# 사이킷런의 LabelEncoder을 사용한 레이블 인코딩
class_le = LabelEncoder()
y = class_le.fit_transform(df['classlabel'].values)
y
```

```python
# 거꾸로 매핑
class_le.inverse_transform(y)
```

```python
# 클래스 레이블을 문자열에서 정수로 바꿉니다
df['classlabel'] = df['classlabel'].map(class_mapping)
df
```

```python
# 클래스 레이블을 거꾸로 매핑합니다
inv_class_mapping = {v: k for k, v in class_mapping.items()}
df['classlabel'] = df['classlabel'].map(inv_class_mapping)
df
```

```python
# 4.3
df_wine = pd.read_csv('https://archive.ics.uci.edu/'
                      'ml/machine-learning-databases/wine/wine.data',
                      header=None)

df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash',
                   'Alcalinity of ash', 'Magnesium', 'Total phenols',
                   'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                   'Color intensity', 'Hue', 'OD280/OD315 of diluted wines',
                   'Proline']

print('Class labels', np.unique(df_wine['Class label']))
df_wine.head()
```

```python
from sklearn.model_selection import train_test_split

X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0,
    stratify=y
    )
```

```python
#4.4
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

X_train_norm = mms.fit_transform(X_train)
X_test_norm = mms.transform(X_test)
```

```python
# 스케일러 생성
mms = MinMaxScaler()
# 학습 (train 데이터로 기준 만들기)
mms.fit(X_train)

# 여기서 내부적으로 min, max 저장됨

# train 변환
X_train_norm = mms.transform(X_train)
# test 변환 (같은 기준 사용)
X_test_norm = mms.transform(X_test)
```

```python
#4.5
import pandas as pd
from sklearn.linear_model import LogisticRegression
# 샘플 데이터 (컬럼 3개)
X = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],   # A와 거의 같은 정보 (불필요 가능성)
    'C': [5, 3, 6, 2, 1]
})

# 타겟
y = [0, 0, 1, 1, 1]
# L1 규제 모델 생성
model = LogisticRegression(penalty='l1', solver='liblinear')
# 학습
model.fit(X, y)
# 각 컬럼의 중요도 (계수)
coef = model.coef_[0]
print("각 feature의 계수:", coef)
# 0이 아닌 컬럼만 선택 (feature selection)
selected_features = X.columns[coef != 0]
print("선택된 feature:", selected_features)
```

```python
from sklearn.linear_model import LogisticRegression
# L2 규제 모델
model = LogisticRegression(penalty='l2')
# 학습
model.fit(X, y)
coef = model.coef_[0]
print("각 feature의 계수:", coef)
```

```python
#4.6
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# 데이터 생성
X = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],   # A와 거의 동일 (중복 정보)
    'C': [5, 3, 6, 2, 1]
})

y = [0, 0, 1, 1, 1]


# 랜덤포레스트 모델 생성
model = RandomForestClassifier(random_state=0)

# 학습
model.fit(X, y)

# feature 중요도 확인
importances = model.feature_importances_

# 보기 쉽게 출력
for col, val in zip(X.columns, importances):
    print(col, ":", val)
```

```python
import numpy as np
# 기준 설정 (예: 0.2 이상만 선택)
threshold = 0.2
selected_features = X.columns[importances > threshold]
print("선택된 feature:", selected_features)
```
