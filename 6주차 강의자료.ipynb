```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Define a custom tokenizer function that includes lemmatization
def custom_tokenizer_with_lemmatization(text):
    # Tokenize the text, convert to lowercase, and lemmatize each token
    tokens = word_tokenize(text.lower())
    lemmatized_tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return lemmatized_tokens

# Initialize CountVectorizer with ngram_range and the custom tokenizer
# Note: When using a custom tokenizer, ngram_range will apply to the tokens returned by the tokenizer.
count_bigrams = CountVectorizer(ngram_range=(1, 2), tokenizer=custom_tokenizer_with_lemmatization)
docs = np.array([
    'The sun is shining',
    'The weather is sweet',
    'The sun is shining, the weather is sweet, and one and one is two'])

# Fit and transform the documents
bag_bigrams = count_bigrams.fit_transform(docs)

print("Vocabulary with bigrams (using custom tokenizer with lemmatization):")
print(count_bigrams.vocabulary_)
print("\nBag of words with bigrams (using custom tokenizer with lemmatization):")
print(bag_bigrams.toarray())
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# X is the bag_bigrams generated from the documents
X = bag_bigrams.toarray()

# Define the target variable y
y = np.array([1, 1, 0])

# Initialize and train a Logistic Regression model
model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(X, y)

print("Logistic Regression Model trained successfully.")
print(f"Model coefficients: {model.coef_}")
print(f"Model intercept: {model.intercept_}")

# Predict on the training data to show how it works
predictions = model.predict(X)
print(f"\nPredictions on training data: {predictions}")
print(f"Actual target values: {y}")
```
