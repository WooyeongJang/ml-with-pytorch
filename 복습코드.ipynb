```python
from sklearn import datasets

# Iris 데이터 로드
iris = datasets.load_iris()

# 처음 100개 샘플만 사용 → Setosa(0) + Versicolor(1)
X = iris.data[:100, [0, 2]]  # Sepal length, Petal length
y = iris.target[:100]
```

```python
import numpy as np


class Perceptron:

    def __init__(self, eta=0.01, n_iter=50, random_seed=1):
        """퍼셉트론 초기 설정"""
        self.eta = eta
        self.n_iter = n_iter
        self.random_seed = random_seed

    def fit(self, X, y):
        """훈련 데이터로 가중치 학습"""
        rgen = np.random.RandomState(self.random_seed)

        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1])
        self.b_ = 0.0
        self.errors_ = []

        for _ in range(self.n_iter):

            errors = 0

            for xi, target in zip(X, y):

                update = self.eta * (target - self.predict(xi))

                self.w_ += update * xi
                self.b_ += update

                errors += int(update != 0.0)

            self.errors_.append(errors)

        return self

    def net_input(self, X):
        """선형 결합"""
        return np.dot(X, self.w_) + self.b_

    def predict(self, X):
        """계단 함수 기반 분류"""
        return np.where(self.net_input(X) >= 0.0, 1, 0)
```

```python

# 모델 생성
ppn = Perceptron(eta=0.01, n_iter=10)

# 학습
ppn.fit(X, y)

# 예측
y_pred = ppn.predict(X)

# 정확도
print(np.mean(y_pred == y))
```

```python
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()

X = iris.data[:100, [0,2]]
y = iris.target[:100]

# label 변환
y = np.where(y == 0, -1, 1)

# scaling
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
import numpy as np


class Adaline:

    def __init__(self, eta=0.01, n_iter=50, random_seed=1):
        """초기 설정"""
        self.eta = eta
        self.n_iter = n_iter
        self.random_seed = random_seed

    def fit(self, X, y):
        """데이터를 이용해 가중치 학습"""
        rgen = np.random.RandomState(self.random_seed)

        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1])
        self.b_ = 0.0
        self.losses_ = []

        for _ in range(self.n_iter):

            net_input = self.net_input(X)
            output = self.activation(net_input)

            errors = y - output

            self.w_ += self.eta * X.T.dot(errors)
            self.b_ += self.eta * errors.sum()

            loss = (errors ** 2).mean() / 2.0
            self.losses_.append(loss)

        return self

    def net_input(self, X):
        """선형 결합"""
        return np.dot(X, self.w_) + self.b_

    def activation(self, z):
        """Adaline 활성화 함수"""
        return z
    def predict(self, X):
        return np.where(self.activation(self.net_input(X)) >= 0.0, 1, -1)
```

```python
# 모델 생성
ada = Adaline(eta=0.01, n_iter=20)

# 학습
ada.fit(X, y)

# 예측
y_pred = ada.predict(X)

# 정확도
print(np.mean(y_pred == y))
```
